```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Ejercicio 1: Serie Fibonacci
n = int(input("Ingrese un número para la serie Fibonacci: "))
def fibonacci(n):
    fib_series = [0, 1]
    for _ in range(n - 2):
        fib_series.append(fib_series[-1] + fib_series[-2])
    return fib_series

def ejercicio_1():
    series = fibonacci(n)
    plt.plot(series, marker='o', linestyle='-')
    plt.title("Serie de Fibonacci")
    plt.xlabel("Índice")
    plt.ylabel("Valor")
    plt.grid()
    plt.show()

ejercicio_1()
```

```python
# Ejercicio 2: Análisis de datos meteorológicos
temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]
def ejercicio_2():
    media = np.mean(temperaturas)
    mediana = np.median(temperaturas)
    desviacion = np.std(temperaturas)
    print(f"Media: {media}, Mediana: {mediana}, Desviación estándar: {desviacion}")
    plt.plot(temperaturas, marker='o', linestyle='-')
    plt.title("Temperatura Media Mensual")
    plt.xlabel("Meses")
    plt.ylabel("Temperatura")
    plt.grid()
    plt.show()

ejercicio_2()
```

```python
# Ejercicio 3: Simulación de un dado
def ejercicio_3():
    lanzamientos = np.random.randint(1, 7, 1000)
    valores, conteo = np.unique(lanzamientos, return_counts=True)
    plt.bar(valores, conteo)
    plt.title("Frecuencia de cada número en 1000 lanzamientos de un dado")
    plt.xlabel("Número del dado")
    plt.ylabel("Frecuencia")
    plt.show()

ejercicio_3()
```

```python
# Ejercicio 4: Gráfica de funciones trigonométricas
def ejercicio_4():
    x = np.linspace(0, 2 * np.pi, 1000)
    plt.figure(figsize=(8, 5))
    plt.plot(x, np.sin(x), label="Seno", linestyle='-', marker='o', markersize=4)
    plt.plot(x, np.cos(x), label="Coseno", linestyle='--', marker='s', markersize=4)
    plt.plot(x, np.tan(x), label="Tangente", linestyle=':', marker='x', markersize=4)
    plt.ylim(-2, 2)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.legend()
    plt.grid()
    plt.title("Funciones Trigonométricas")
    plt.show()

ejercicio_4()
```

```python
# Ejercicio 5: Raíces cuadradas
def ejercicio_5():
    x = np.linspace(0, 10, 700)
    plt.plot(x, np.sqrt(x), label="y = sqrt(x)")
    plt.plot(x, x, label="y = x")
    plt.legend()
    plt.title("Raíces Cuadradas")
    plt.show()

ejercicio_5()
```

```python
# Ejercicio 6: Pendiente de una recta
x1, y1 = map(float, input("Ingrese x1 y y1 separados por espacio: ").split())
x2, y2 = map(float, input("Ingrese x2 y y2 separados por espacio: ").split())
def ejercicio_6():
    if x1 == x2:
        print("La recta es vertical y no puede ser graficada correctamente.")
        return
    m = (y2 - y1) / (x2 - x1)
    b = y1 - m * x1
    x = np.linspace(min(x1, x2), max(x1, x2), 100)
    y = m * x + b
    plt.plot(x, y, label=f"Pendiente = {m:.2f}")
    plt.scatter([x1, x2], [y1, y2], color='red')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.legend()
    plt.grid()
    plt.title("Pendiente de una recta")
    plt.show()

ejercicio_6()
```

```python
# Ejercicio 7: Movimiento parabólico
v0 = float(input("Ingrese velocidad inicial: "))
angulo = float(input("Ingrese ángulo de lanzamiento (grados): "))
def ejercicio_7():
    g = 9.81
    t = np.linspace(0, 2 * v0 * np.sin(np.radians(angulo)) / g, num=100)
    x = v0 * np.cos(np.radians(angulo)) * t
    y = v0 * np.sin(np.radians(angulo)) * t - 0.5 * g * t**2
    plt.plot(x, y)
    plt.title("Movimiento Parabólico")
    plt.xlabel("Distancia")
    plt.ylabel("Altura")
    plt.show()

ejercicio_7()
```

```python
# Ejercicio 8: Funciones potencia
n = int(input("Ingrese el valor de n para la función potencia: "))
def ejercicio_8():
    x = np.linspace(1, 10, 100)
    y = x ** n
    plt.plot(x, y, label=f"y = x^{n}")
    plt.legend()
    plt.title("Funciones Potencia")
    plt.show()

ejercicio_8()
```

```python
# Ejercicio 9: Fractal de Sierpinski
def ejercicio_9():
    puntos = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3) / 2]])
    p = np.array([0.5, 0.5])
    iteraciones = 60000
    x, y = [], []
    while iteraciones > 0:
        p = (p + random.choice(puntos)) / 2
        x.append(p[0])
        y.append(p[1])
        iteraciones -= 1
    plt.scatter(x, y, s=0.2, color='black')
    plt.title("Fractal de Sierpinski")
    plt.axis("off")
    plt.show()

ejercicio_9()
```

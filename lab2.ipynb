def newton(fun, der, x_n, epsilon=1e-6, steps=50):  #  Define la función de Newton-Raphson con sus parámetros.
    for n in range(steps + 1):                      #  Itera hasta un máximo de 'steps' veces.
        f_x = fun(x_n)                              # Evalúa la función en el punto actual.
        if abs(f_x) < epsilon:                      # Si el valor absoluto de la función es suficientemente pequeño...
            return x_n                              # ... entonces se retorna x_n como la raíz encontrada.

        d_f = der(x_n)                              # Evalúa la derivada en el punto actual.
        if d_f == 0:                                # Si la derivada es cero
            print('Error la derivada es cero')      # se muestra un error
            return None                             # y se detiene el algoritmo.

        x_n = x_n - f_x / d_f                       # Actualiza el valor de x_n usando la fórmula de Newton-Raphson.

    return None                                     # Si no se encuentra una raíz en 'steps' iteraciones, retorna None.


```python

#Ejercicio 3
#3.1
import numpy as np
import matplotlib.pyplot as plt

# Método de Newton-Raphson
def newton_raphson(f, fprime, x0, tol=1e-4, maxiter=50):
    x = x0
    for i in range(maxiter):
        fx = f(x)
        if abs(fx) < tol:
            return x
        fp = fprime(x)
        if fp == 0:
            print('Error: la derivada es cero')
            break
        x = x - fx / fp
    return x

# Funciones y derivadas
f1 = lambda x: x*5 - 6*x*3 + 2*x - 2
df1 = lambda x: 5*x*4 - 18*x*2 + 2

f2 = lambda x: np.sin(x) - x**2
df2 = lambda x: np.cos(x) - 2*x

f3 = lambda t: np.cos(t) - t**2
df3 = lambda t: -np.sin(t) - 2*t

f4 = lambda x: 2*np.sin(x) - np.sqrt(x + 1)
df4 = lambda x: 2*np.cos(x) - 1/(2*np.sqrt(x + 1))

# Graficar funciones
x_vals = np.linspace(-3, 3, 400)
functions = [(f1, "f1(x)"), (f2, "f2(x)"), (f3, "f3(t)"), (f4, "f4(x)")]

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs = axs.flatten()
for i, (f, label) in enumerate(functions):
    axs[i].axhline(0, color='black', linewidth=0.5)
    axs[i].plot(x_vals, f(x_vals), label=label)
    axs[i].legend()
    axs[i].grid(True)
plt.tight_layout()
plt.show()

# Aproximaciones iniciales obtenidas del gráfico
initial_guesses = [0.5, 0.5, 0.8, 1.2]
functions_and_derivatives = [(f1, df1), (f2, df2), (f3, df3), (f4, df4)]

# Ejecutar Newton-Raphson
solutions = []
for (f, df), x0 in zip(functions_and_derivatives, initial_guesses):
    root = newton_raphson(f, df, x0, tol=1e-4)
    solutions.append(root)

# Mostrar resultados
for i, sol in enumerate(solutions, start=1):
    print(f"Solución {i}: {sol}")


#Ejercio 3.2
# Definimos la función y su derivada
def f(x):
    return x**3 - 2

def f_prime(x):
    return 3 * x**2

# Método de Newton-Raphson
def newton_raphson(f, fprime, x0, tol, maxiter):
    x = x0
    for i in range(maxiter):
        fx = f(x)
        if abs(fx) < tol:
            print(f"Iteraciones: {i}")
            return x  # Solución encontrada con tolerancia deseada

        fp = fprime(x)
        if fp == 0:
            print("Error: la derivada es cero")
            return None

        x = x - fx / fp  # Actualización de Newton

    print("No se alcanzó la tolerancia en el número máximo de iteraciones")
    return x

# Parámetros iniciales
x0 = 1.0          # Suposición inicial
tol = 0.00005     # Tolerancia deseada
maxiter = 100     # Máximo número de iteraciones

# Llamada al método de Newton-Raphson
resultado = newton_raphson(f, f_prime, x0, tol, maxiter)

# Imprimir resultado
if resultado is not None:
    print(f"Raíz cúbica aproximada de 2: {resultado}")
    print(f"Error absoluto: {abs(resultado**3 - 2)}")

```
